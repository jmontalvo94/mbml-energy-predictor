```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
datadir = '/'.join(os.getcwd().split('/')[:-1])+"/2_Data/"
```

```python
meta_df = pd.read_csv(datadir+'building_metadata.csv')
meta_df.head()
print(meta_df.shape)
#meta_df = meta_df.dropna()
#print(meta_df.shape)
```

```python
len(set(meta_df.building_id))
```

```python
plt.figure(figsize=(16,7))
sns.countplot(meta_df.primary_use, alpha=0.8, label='Full')
plt.xticks(rotation=15)
```

```python
plt.figure(figsize=(16,7))
sns.countplot(meta_df.primary_use,color='g', alpha=0.5, label='Full')
plt.xticks(rotation=15)

sns.countplot(meta_df.dropna().primary_use,color = 'r', alpha=0.8, label='After dropping NaNs')

plt.xticks(rotation=15)
plt.legend()
```

```python
# missing values per category
meta_df.set_index('primary_use').isna().mean(level=0)
```

```python
meta_df.groupby('primary_use').mean()
```
